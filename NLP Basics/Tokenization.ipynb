```python
corpus = """Hello Welcome, to Sai Srushik.
Please do watch the entire course! to become an expert
"""
```

```python
import nltk
nltk.download('punkt_tab')
```

```python
#Tokenization

from nltk.tokenize import sent_tokenize, word_tokenize

sent_tokenize(corpus)
```

```python
documents = sent_tokenize(corpus)
type(documents)
```

```python
for sentence in documents:
    print(sentence)
```

```python
#Word Tokenization
#Paragraphs --> words
word_tokenize(corpus)
```

```python
#sentence--> words
res = []
for sentence in documents:
    res.append(word_tokenize(sentence))
print(res)
```

```python
from nltk.tokenize import wordpunct_tokenize
#pucntuations are also created as words
wordpunct_tokenize(corpus)
```

```python
from nltk.tokenize import TreebankWordTokenizer
```

```python
tokenizer = TreebankWordTokenizer()
#Full stop cannot be considered as a separate word
tokenizer.tokenize(corpus)
```
